## Missing values

```python
# missing values
df.isna().sum()
```

```python
# Try to replace Null values
df['Product_Category_2'].unique()         # This feature will only become Discrete data beacuse it will only repeat it
```

####     df['Product_Category_2'].unique()
###### The feature 'Product_Category_2' will become discrete because the .unique() method is used to extract the unique values present in that specific column. Discrete data consists of distinct, separate values without any meaningful values in between. When you use .unique(), you are essentially identifying and listing all the distinct categories or values within that column, which typically represent specific, non-continuous categories or classes. This is in contrast to continuous data, which can take on a wide range of values within a certain range.

```python
# Check the values how many time it is repeating
df['Product_Category_2'].value_counts()
```

#### Here the best way to remove the discrete values or Continous Values is 'Mode'. Here We can't use 'Mean' because it will create a new category.

```python
df['Product_Category_2'].mode()
```

```python
# Remove null values by mode 
df['Product_Category_2']=df['Product_Category_2'].fillna(df['Product_Category_2'].mode()[0])
```

#### Now Let's Checkout  why we use mode()[0] here 
##### df['Product_Category_2'].mode()[0]

###  OBSERVATION 

##### Based on the provided graphs, it's evident that males tend to make more purchases than females.

```python
# Create data for the pie chart
gender_counts = df['Gender'].value_counts()

# Define labels and sizes
labels = gender_counts.index
sizes = gender_counts.values

# Define colors
colors = ['crimson', 'pink']

# Explode a piece of the pie
explode = (0.1, 0)  # The first slice will be exploded

# Create a pie chart
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90, explode=explode)

# Set aspect ratio to be equal, so the pie will be circular
plt.axis('equal')

# Display the pie chart
plt.show()
```
